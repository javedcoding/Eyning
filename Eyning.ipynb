This module will have the eye training of 9 places and words can be in any of the 9 places randomly each time. 

```python
#First we need to import all the libraries
from tkinter import *
from tkinter import messagebox
from PyPDF2 import PdfFileReader
import time
import random

#Initialize the window 
window = Tk()
window.minsize(700, 700)
window.title('Eyning (Train your eyes)')
#window.iconbitmap('icon.bmp')

stop_called = False

#this is the main funtion where everything is processed upon calling start button
def start_process():
    '''Thiw is the main function for wiring up of gui and back end functions'''
    global stop_called
    stop_called = False
    print('start_process being called')
    stop_button.config(state=ACTIVE)
    start_button.config(state=DISABLED)
    
    #now let's break the words first
    words = []
    with open('EyeTraining.txt', 'r', encoding='utf8') as text:
        for line in text:
            for word in line.split():
                words.append(word)
    read_word_label_5.config(text="")
    
    #Here we will wire up all the widgets and calculate seconds required for a word
    wpm = int(wpm_text.get()) #wiring up the word per minute to calculate
    time_for_aword = 60 / wpm
    prev_place = 0
    
    for i in words:
        if stop_called == False:
            random_place = random.randint(1,9)
            if random_place == prev_place:
                random_place = random.randint(1,9)
            
            label_text = "read_word_label_"+str(random_place)
            label = eval(label_text)
            label.config(text=i)
            label.update()
            time.sleep(time_for_aword)
            prev_place = random_place
            label.config(text="")
            label.update()
        else:
            break
    stop_button.config(state=DISABLED)
    start_button.config(state=ACTIVE) 
    read_word_label_5.config(text="Read Here")
    read_word_label_5.update()

#this is the stop button funtion
def stop_process():
    print('stop_process being called')
    global stop_called 
    stop_called = True
    read_word_label_1.config(text="")
    read_word_label_1.update()
    read_word_label_2.config(text="")
    read_word_label_2.update()
    read_word_label_3.config(text="")
    read_word_label_3.update()
    read_word_label_4.config(text="")
    read_word_label_4.update()
    read_word_label_5.config(text="Read Here")
    read_word_label_5.update()
    read_word_label_6.config(text="")
    read_word_label_6.update()
    read_word_label_7.config(text="")
    read_word_label_7.update()
    read_word_label_8.config(text="")
    read_word_label_8.update()
    read_word_label_9.config(text="")
    read_word_label_9.update()
    stop_button.config(state=DISABLED)
    start_button.config(state=ACTIVE)
    
#this is the function for closing the windows
def on_closing():
    if messagebox.askokcancel("Quit", "Do you want to quit?"):
        window.destroy() #this will destroy the tkinter window

#Words Per Minute
wpm_text = StringVar()
wpm_label = Label(window, text='Please give desired word per minute in integer ', width=20, font=('bold', 10), pady=10, wraplength=200)
wpm_label.grid(row=0, column=0, sticky=W)
wpm_entry = Entry(window, textvariable=wpm_text, width=10)
wpm_entry.grid(row=0, column=1)

#Start and Stop Button
start_button = Button(window, text="Start", state=ACTIVE, command=start_process)
start_button.grid(row=1, column=1)
stop_button = Button(window, text="Stop", state=DISABLED, command=stop_process)
stop_button.grid(row=1, column=2)

#Word Showing Label1
read_word_label_1 = Label(window, text="", font=('bold', 15), width=20, height=8, padx=1, pady=1, bg='#ffffff')
read_word_label_1.grid(row=5, column=0)
read_word_label_2 = Label(window, text="", font=('bold', 15), width=20, height=8, padx=1, pady=1, bg='#ffffff')
read_word_label_2.grid(row=5, column=1)
read_word_label_3 = Label(window, text="", font=('bold', 15), width=20, height=8, padx=1, pady=1, bg='#ffffff')
read_word_label_3.grid(row=5, column=2)
read_word_label_4 = Label(window, text="", font=('bold', 15), width=20, height=8, padx=1, pady=1, bg='#ffffff')
read_word_label_4.grid(row=6, column=0)
read_word_label_5 = Label(window, text="Read Here", font=('bold', 15), width=20, height=8, padx=1, pady=1, bg='#ffffff')
read_word_label_5.grid(row=6, column=1)
read_word_label_6 = Label(window, text="", font=('bold', 15), width=20, height=8, padx=1, pady=1, bg='#ffffff')
read_word_label_6.grid(row=6, column=2)
read_word_label_7 = Label(window, text="", font=('bold', 15), width=20, height=8, padx=1, pady=1, bg='#ffffff')
read_word_label_7.grid(row=7, column=0)
read_word_label_8 = Label(window, text="", font=('bold', 15), width=20, height=8, padx=1, pady=1, bg='#ffffff')
read_word_label_8.grid(row=7, column=1)
read_word_label_9 = Label(window, text="", font=('bold', 15), width=20, height=8, padx=1, pady=1, bg='#ffffff')
read_word_label_9.grid(row=7, column=2)

window.protocol("WM_DELETE_WINDOW", on_closing) #this will happen on close sign clicking
window.mainloop()
```
